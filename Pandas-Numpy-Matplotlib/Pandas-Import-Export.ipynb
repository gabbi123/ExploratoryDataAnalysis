```python
import pandas as pd
```

```python
# read csv and write to csv
cars_csv = pd.read_csv('Data/Cars2015.csv')
# print(cars_csv.head(10))
# print(cars_csv.groupby('Make')['HwyMPG'].mean())
make_string_lengths = cars_csv['Make'].unique()
top_5_make_names = pd.Series(make_string_lengths).str.len()

result = pd.DataFrame({'Make': make_string_lengths, 'Name_Length': top_5_make_names})
result = result.sort_values('Name_Length', ascending=False).head(5)

print(result)
```

```python
cars_csv.to_csv('Data/Cars2015-to-csv.csv')
```
