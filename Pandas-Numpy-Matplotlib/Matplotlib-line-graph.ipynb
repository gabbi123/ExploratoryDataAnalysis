```python
# Import Matplotlib submodule for plotting
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([1,2,3,4]) # List of vertical co-ordinates of the points plotted
plt.show() # Displays plot
# Implicit X-axis values from 0 to (N-1) where N is the length of the list
```

```python
# Specify value of both X and Y
x = range(5)

plt.plot(x, [ x1*2 for x1 in x ]) # (x, y)
plt.show()
```

```python
# use numpy to define the valules
x = np.arange(0, 5, 0.1)
plt.plot(x, [x1**2 for x1 in x])
plt.show()
```

```python
# multiline plot
x = range(5)
plt.plot(x, [x1 for x1 in x])
plt.plot(x, [x1 * x1 for x1 in x])
plt.plot(x, [x1 * x1 * x1 for x1 in x])
plt.show()
```

```python
# inline multiplot
x = range(5)
plt.plot(x, [x1 for x1 in x],
         x, [x1 * x1 for x1 in x],
         x, [x1 * x1 * x1 for x1 in x])
plt.show()
```

```python
# add grid lines
x = range(5)

plt.plot(x, [x1 for x1 in x],
         x, [x1 * x1 for x1 in x],
         x, [x1 * x1 * x1 for x1 in x])
plt.grid(True)
plt.show()
```

```python
# set axis limits
x = range(5)

plt.plot(x, [x1 for x1 in x],
         x, [x1 * x1 for x1 in x],
         x, [x1 * x1 * x1 for x1 in x])
plt.grid(True)
plt.axis([-1, 5, -1, 10])
plt.show()
```

```python
# set axis limits using xlim and ylim
x = range(5)

plt.plot(x, [x1 for x1 in x],
         x, [x1 * x1 for x1 in x],
         x, [x1 * x1 * x1 for x1 in x])
plt.grid(True)
plt.xlim(-1, 5)
plt.ylim(-1, 10)
plt.show()
```

```python
# add labels for axis
x = range(5)

plt.plot(x, [x1 for x1 in x],
         x, [x1 * x1 for x1 in x],
         x, [x1 * x1 * x1 for x1 in x])
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
# add plot title
x = range(5)

plt.plot(x, [x1 for x1 in x],
         x, [x1 * x1 for x1 in x],
         x, [x1 * x1 * x1 for x1 in x])
plt.grid(True)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sample line graph')
plt.show()
```

```python
# add legend
x = range(5)

plt.plot(x, [x1 for x1 in x], label = 'linear')
plt.plot(x, [x1 * x1 for x1 in x], label = 'square')
plt.plot(x, [x1 * x1 * x1 for x1 in x], label = 'cube')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sample line graph')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# add markers
x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]

plt.plot(x, y, 'bo')
for i in range(len(x)):
    x_plot = x[i]
    y_plot = y[i]
    plt.text(x_plot, y_plot, (x_plot, y_plot), fontsize = 10)

plt.show()
```

```python
# save plot
x = np.arange(1, 10, 0.1)

plt.plot(x, [x1 for x1 in x], label = 'linear')
plt.plot(x, [x1 * x1 for x1 in x], label = 'square')
plt.plot(x, [x1 * x1 * x1 for x1 in x], label = 'cube')

plt.xlabel = 'X-axis'
plt.ylabel = 'Y-axis'
plt.legend()
plt.grid(True)
plt.title = 'Sample plot'

plt.savefig('Data/sampleplot.png')
plt.show()
```
