```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Histogram
y = np.random.randn(100, 100)
plt.hist(y)
plt.show()
```

```python
# use bins to make it non overlapping. Default bin value is 10
y = np.random.randn(1000)
plt.hist(y,100)
plt.show()
```

```python
# Bar chart
plt.bar([1,2,3],[2,4,6])  # midpoint of bar, value of bar
plt.show()
```

```python
dct = {'A': 1, 'B': 3, 'C': 5}
for index, key in enumerate(dct):
    plt.bar(index, dct[key]) # dictionaries are not iterable by default hence ploting individually

plt.show()
```

```python
# display x-label
dct1 = {'A': 10, 'B': 20, 'C': 30 }
for index, key in enumerate(dct1):
    plt.bar(index, dct1[key])

plt.xticks(np.arange(len(dct1)), dct1.keys())
plt.show()
```
