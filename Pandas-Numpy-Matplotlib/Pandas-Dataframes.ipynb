```python
import pandas as pd
```

```python
# create df from list
list1 = [1,2,3,4]
table = pd.DataFrame(list1)
print (table)
```

```python
# create df from dictionary
dct1 = [{'a': 1, 'd': 2}, {'a': 1, 'b': 2, 'c': 3}]
df1 = pd.DataFrame(dct1)
print(df1)
```

```python
# create df with index
df2 = pd.DataFrame(dct1, index = ['First', 'Second'])
print (df2)
```

```python
# dictionary of series to dataframe
dct3 = {'one': pd.Series([1,2,3], index = ['a', 'b','c']),
        'two': pd.Series([1,2,3,4], index = ['a', 'b', 'c', 'd'])}
df3 = pd.DataFrame(dct3)
print(df3)
```

```python
# a new column added
df3['three'] = pd.Series([2,3], index = ['c', 'd'])
print(df3)
```

```python
# delete df column
del df3['one']
print(df3)
```

```python
# delete df column using pop
df3.pop('three')
print(df3)
```

```python
# fetch row using row label
print(df3.loc['c'])
```

```python
# fetch row using index
print(df3.iloc[1])
```

```python
# append row function
data2 = {'one':pd.Series([1,2,3], index = ['a', 'b', 'c']),
        'two':pd.Series([1,2,3,4], index = ['a', 'b', 'c', 'd'])}
table5 = pd.DataFrame(data2)
table5['three'] = pd.Series([10,20,30], index = ['a', 'b', 'c'])
row = pd.DataFrame([[11,13],[17,19]], columns = ['two', 'three'])
table6 = table5.append(row)
print(table6)
```

```python
# drop rows
table6 = table5.drop('b')
print(table6)
```
