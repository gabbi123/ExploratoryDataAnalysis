```python
import numpy as np
```

```python
# create one dimensional array
arr = np.array([1,2,3])
print(arr)
```

```python
# multi dimensional array
multiarr = np.array([[1,2],[3,4]])
print (multiarr)
```

```python
# list
list1 = [1,2,3,4,5]
lsarr = np.array(list1)
print(lsarr)
```

```python
# array with specified range
numarr = np.arange(5,50)
print(numarr)
```

```python
# 5x5 array with zeros
fivearr = np.zeros((5,5))
print(fivearr)
```

```python
# create linearly spaced array
linarr = np.linspace(0, 20, 5)
print(linarr)
```

```python
# create array from existing data
arr1 = [1, 2, 3]
toarr = np.asarray(arr1)
print(toarr)
```
