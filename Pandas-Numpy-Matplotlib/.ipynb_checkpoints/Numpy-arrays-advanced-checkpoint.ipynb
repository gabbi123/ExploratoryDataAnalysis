```python
import numpy as np
```

```python
# array indexing
nparr = np.arange(2, 20)
print(nparr[1])
print(nparr.size)
```

```python
# slice array
arr1 = np.arange(1, 20)
arr_slice = slice(1, 20, 2)

print(arr1[arr_slice])
```

```python
# slice array inline
arr2 = np.arange(1, 20)
print(arr2[2:])
```

```python
# slice until specified index
print(arr2[:15])
```

```python
# slicing of 2d array
arr3 = np.array([[1,2,3],[3,4,5],[4,5,6]])
print(arr3[:2, :2])
```

```python
# array attributes
print(arr3.shape)
print(arr3.ndim)
print(arr3.size)
print(arr3.itemsize)
```

```python
# create uninitialized array
arr4 = np.empty([3,2], dtype=int)
print(arr4)
```

```python
# create 1D array with zeros
arr5 = np.zeros(5)
print(arr5)
```
