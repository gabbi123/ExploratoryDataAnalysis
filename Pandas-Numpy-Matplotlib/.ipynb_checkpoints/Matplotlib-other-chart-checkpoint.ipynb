```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# pie chart
plt.figure(figsize = (3,3))
x = [40, 20, 5]
labels = ['Bike', 'Car', 'Truck']
plt.pie(x, labels = labels)
plt.show()
```

```python
# scatter plot
x = np.random.rand(100)
y = np.random.rand(100)

plt.scatter(x,y)
plt.show()
```

```python
# custom line colors
x = np.arange(1, 3, 0.5)
plt.plot(x, 'y')
plt.plot(x * 2, 'm')
plt.plot(x * 4, 'b')

plt.show()

#
#Color code:

# b = Blue
# c = Cyan
# g = Green
# k = Black
# m = Magenta
# r = Red
# w = White
# y = Yellow
#
```

```python
# different line styles for the plot
# - Solid line
# -- Dashed line
# -. Dash-Dot line
# : Dotted Line

y = np.arange(1, 100)
plt.plot(y, '--',
         y + 10, '-.',
         y + 20, ':')
plt.show()
```

```python
# custom option for markers
y = np.arange(1, 5, 0.1)

plt.plot(y, '*',
         y + 10, '^',
         y + 20, 'D',
         y + 30, 'o')
plt.show()
```
