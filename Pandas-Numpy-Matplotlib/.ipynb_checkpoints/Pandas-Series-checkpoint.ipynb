```python
import pandas as pd
import numpy as np
```

```python
# create series
series = pd.Series()
print(series)
```

```python
# create series from ndarray
arr = np.array([1,2,3,4,5])
series1 = pd.Series(arr)
print(series1)
```

```python
# create series from python dictionary
data = {'a': 10, 'b':20, 'c':30}
series2 = pd.Series(data)
print(series2)
```

```python
# slice pd array
print(series1[0:3])
```
